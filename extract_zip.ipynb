```python
import glob
import zipfile
import os
import shutil
```

```python
# Define the directory containing the ZIP files
zip_dir = os.getcwd()
# Define the target directory for the extracted file
target_dir = os.path.join(os.getcwd(), 'US_Text')
# Define a pattern or specific filename you're looking for within the ZIP files
#target_file_pattern = ['us', 'US'] 

# Find all ZIP files in the directory
zip_files = glob.glob(os.path.join(zip_dir, '*.zip'))

for zip_file in zip_files:
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        # Extract all files to a temporary directory (or you can choose to extract only the one you need if you know its name)
        zip_ref.extractall('/tmp/zip_extraction')
        
        # Iterate through the extracted files to find the one you want
        for extracted_file in zip_ref.namelist():
            if 'us' in extracted_file or 'US' in extracted_file:
                # Construct the full path to the extracted file
                extracted_file_path = os.path.join('/tmp/zip_extraction', extracted_file)
                # Construct the full target path
                target_file_path = os.path.join(target_dir, os.path.basename(extracted_file))
                
                # Copy the file to the target directory
                shutil.copy(extracted_file_path, target_file_path)
                print(f'Copied {extracted_file} to {target_dir}')
                
        # Optionally, clean up the temporary extraction directory
        shutil.rmtree('/tmp/zip_extraction')
```
